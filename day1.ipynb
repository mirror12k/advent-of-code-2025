```perl
#!/usr/bin/env perl
use strict;
use warnings;
use feature qw/ say current_sub state /;
use Data::Dumper;
use lib '.';
use toolkit;
use graphql;
use adventofcode;

$Data::Dumper::Sortkeys = 1;
```

```perl
my $practice_input = qq/
L68
L30
R48
L5
R60
L55
L1
L99
R14
L82
/;
my $practice_input2 = qq/
R1000
L1000
R100
L100
/;
```

```perl

sub process_day1 {
    my ($in) = @_;
    my $dial = 50;
    my @steps = ($dial);
    my @larger_steps;
    my $total_part1 = 0;
    my $total_part2 = 0;
    for my $rot (map s/R//r, map s/L/-/r, grep $_, split '\n', $in) {
        timecheck;

        my $res = $dial + $rot;
        if ($rot < 0) {
            push @steps, map $_ % 100, reverse $res .. $dial-1;
        } else {
            push @steps, map $_ % 100, $dial+1 .. $res;
        }
        

        $dial += $rot;
        $dial %= 100;
        push @larger_steps, $dial;
    }

    $total_part1 = sum map 1, grep $_ == 0, @larger_steps;
    $total_part2 = sum map 1, grep $_ == 0, @steps;
    # @steps;
    $total_part1, $total_part2;
}

timer sub {
    say join ',', process_day1($practice_input);
    say join ',', process_day1($practice_input2);
}
```
