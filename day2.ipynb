```perl
sub process_day2 {
    my ($input) = @_;
    my @dups = map { my ($start, $end) = split '-'; timecheck; grep /\A(\d+)\1+\Z/, $start .. $end } grep $_, map s/\s*//gr, split ',', $input;
    my @dups_part1 = grep /\A(\d+)\1\Z/, @dups;

    say 'debug p1:', join ',', @dups_part1[0 .. 4];
    say 'debug p2:', join ',', @dups[0 .. 4];

    my $part1_sum = sum @dups_part1;
    my $part2_sum = sum @dups;
    return $part1_sum, $part2_sum;
}

timer sub {
    my ($part1, $part2) = process_day2($practice_input);
    confirm '1227775554', $part1;
    confirm '4174379265', $part2;
}
```

```perl
my ($res1, $res2);
timer sub {
    ($res1, $res2) = process_day2($input);
    say $res2;
    confirm '31210613313', $res1;
    confirm '41823587546', $res2;
}
```
