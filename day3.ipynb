```perl
#!/usr/bin/env perl
use strict;
use warnings;
use feature qw/ say current_sub state /;
use Data::Dumper;
use lib '.';
use toolkit;
use graphql;
use adventofcode;

$Data::Dumper::Sortkeys = 1;
```

```perl
my $practice_input = qq/
987654321111111
811111111111119
234234234234278
818181911112111
/;
```

```perl
my $input = get_challenge('2025/day/3/input');
```

```perl
sub process_day3 {
    my ($input) = @_;
    my @rows = grep $_, map s/\s+//gr, split /\s+/, $input;
    my @tops;
    for my $row (@rows) {
        timecheck;
        my $top = (reverse sort split '', substr $row, 0, -1)[0];
        # say "top: $top";
        # say "search space: ", substr $row, 1+index $row, $top;
        my $second_top = (reverse sort split '', substr $row, 1+index $row, $top)[0];
        # say "second_top: $second_top";
        push @tops, "$top$second_top";
    }
    say 'debug: ', join ', ', @tops;

    sum @tops
}

timer sub {
    say process_day3($practice_input);
    confirm '357', process_day3($practice_input);
}
```

```perl
my $res1;
timer sub {
    ($res1) = process_day3($input);
    say $res1;
    confirm '17107', $res1;
}
```

```perl
say post_answer('2025/day/3/answer', 1, $res1); # 17107
```

```perl
sub find_top {
    my ($s, $requires) = @_;
    my $top_value = (reverse sort split '', substr $s, 0, length ($s)-$requires)[0];
    return index $s, $top_value
}
sub process_day3_part2 {
    my ($input) = @_;
    my @rows = grep $_, map s/\s+//gr, split /\s+/, $input;
    my @tops;
    for my $row (@rows) {
        timecheck;
        my @bs;
        push @bs, find_top($row, 11);
        push @bs, 1 + $bs[-1] + find_top(substr($row, 1+$bs[-1]), 10);
        push @bs, 1 + $bs[-1] + find_top(substr($row, 1+$bs[-1]), 9);
        push @bs, 1 + $bs[-1] + find_top(substr($row, 1+$bs[-1]), 8);
        push @bs, 1 + $bs[-1] + find_top(substr($row, 1+$bs[-1]), 7);
        push @bs, 1 + $bs[-1] + find_top(substr($row, 1+$bs[-1]), 6);
        push @bs, 1 + $bs[-1] + find_top(substr($row, 1+$bs[-1]), 5);
        push @bs, 1 + $bs[-1] + find_top(substr($row, 1+$bs[-1]), 4);
        push @bs, 1 + $bs[-1] + find_top(substr($row, 1+$bs[-1]), 3);
        push @bs, 1 + $bs[-1] + find_top(substr($row, 1+$bs[-1]), 2);
        push @bs, 1 + $bs[-1] + find_top(substr($row, 1+$bs[-1]), 1);
        push @bs, 1 + $bs[-1] + find_top(substr($row, 1+$bs[-1]), 0);
        
        
        # say "second_top: $second_top";
        push @tops, join '', map { substr($row, $_, 1) } @bs
    }
    say 'debug: ', join ', ', @tops;

    sum @tops
}

timer sub {
    say process_day3_part2($practice_input);
    confirm '3121910778619', process_day3_part2($practice_input);
}
```

```perl
my $res2;
timer sub {
    $res2 = process_day3_part2($input);
    say $res2;
    confirm '169349762274117', $res2;
}
```

```perl
say post_answer('2025/day/3/answer', 2, $res2); # 169349762274117
```
