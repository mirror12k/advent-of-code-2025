```perl
#!/usr/bin/env perl
use strict;
use warnings;
use feature qw/ say current_sub state /;
use Data::Dumper;
use lib '.';
use toolkit;
use graphql;
use adventofcode;

$Data::Dumper::Sortkeys = 1;
```

```perl
my $practice_input = qq/
..@@.@@@@.
@@@.@.@.@@
@@@@@.@.@@
@.@@@@..@.
@@.@@@@.@@
.@@@@@@@.@
.@.@.@.@@@
@.@@@.@@@@
.@@@@@@@@.
@.@.@@@.@.
/;
```

```perl
my $input = get_challenge('2025/day/4/input');
```

```perl
sub process_day4 {
    my ($input) = @_;
    my $map =
    # say
        # Dumper
        string_2d_map_array
        # map_nd selector('neighbor.value'),
        map_nd sub { $_->{sum} < 4 && $_->{value} eq '@' ? 'x' : $_->{value} },
        map_nd sub { return {
            value => $_->{value},
            sum => sum map { $_->{where}{value} ? 1 : 0 } @{$_}{qw/ n00 n01 n02 n12 n10 n20 n21 n22 /} } },
        # map_nd selector_multi('n00.value,n01.value,n02.value,n12.value,n10.value,n20.value,n21.value,n22.value'),
        map_nd_indexed graphql_query(q'{
            value
            n00:neighbor(da:-1, db:-1) {where(value:"@") {value}}
            n01:neighbor(da:-1, db:0) {where(value:"@") {value}}
            n02:neighbor(da:-1, db:1) {where(value:"@") {value}}
            n12:neighbor(da:0, db:1) {where(value:"@") {value}}
            n10:neighbor(da:0, db:-1) {where(value:"@") {value}}
            n20:neighbor(da:1, db:-1) {where(value:"@") {value}}
            n21:neighbor(da:1, db:0) {where(value:"@") {value}}
            n22:neighbor(da:1, db:1) {where(value:"@") {value}}
        }'),
        parse_2d_map_array($input);
    say "debug:\n", $map;
    return length $map =~ s/[^x]//gr;
}

timer sub {
    say process_day4($practice_input);
    confirm 13, process_day4($practice_input);
}
```

```perl
my ($res1) = @_;
timer sub {
    $res1 = process_day4($input);
    say "$res1";
    # confirm 13, process_day4($practice_input);
}
```

```perl
say post_answer('2025/day/4/answer', 1, $res1); # 1467
```

```perl
sub iterate_removable_neighbors {
    my ($input) = @_;
    return 
    # say
        # Dumper
        string_2d_map_array
        # map_nd selector('neighbor.value'),
        map_nd sub { $_->{sum} < 4 && $_->{value} eq '@' ? 'x' : $_->{value} },
        map_nd sub { return {
            value => $_->{value},
            sum => sum map { $_->{where}{value} ? 1 : 0 } @{$_}{qw/ n00 n01 n02 n12 n10 n20 n21 n22 /} } },
        # map_nd selector_multi('n00.value,n01.value,n02.value,n12.value,n10.value,n20.value,n21.value,n22.value'),
        map_nd_indexed graphql_query(q'{
            value
            n00:neighbor(da:-1, db:-1) {where(value:"@") {value}}
            n01:neighbor(da:-1, db:0) {where(value:"@") {value}}
            n02:neighbor(da:-1, db:1) {where(value:"@") {value}}
            n12:neighbor(da:0, db:1) {where(value:"@") {value}}
            n10:neighbor(da:0, db:-1) {where(value:"@") {value}}
            n20:neighbor(da:1, db:-1) {where(value:"@") {value}}
            n21:neighbor(da:1, db:0) {where(value:"@") {value}}
            n22:neighbor(da:1, db:1) {where(value:"@") {value}}
        }'),
        parse_2d_map_array($input);
}

sub process_day4_part2 {
    my ($input) = @_;
    my $map = $input;
    $map = iterate_removable_neighbors($map);
    my $total = 0;
    while ($map =~ /x/) {
        timecheck 240;
        my $count = length $map =~ s/[^x]//gr;
        $total += $count;
        $map =~ s/x/./g;
        say "debug: removed $count items";

        $map = iterate_removable_neighbors($map);
    }
    say "debug:\n", $map;
    return $total;
}

timer sub {
    say process_day4_part2($practice_input);
    confirm 43, process_day4_part2($practice_input);
}
```

```perl
my ($res2) = @_;
timer sub {
    $res2 = process_day4_part2($input);
    say "$res2";
    # confirm 13, process_day4($practice_input);
}
```

```perl
say post_answer('2025/day/4/answer', 2, $res2); # 8484
```
