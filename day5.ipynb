```perl
#!/usr/bin/env perl
use strict;
use warnings;
use feature qw/ say current_sub state /;
use Data::Dumper;
use lib '.';
use toolkit;
use graphql;
use adventofcode;

$Data::Dumper::Sortkeys = 1;
```

```perl
my $practice_input = qq/
3-5
10-14
16-20
12-18

1
5
8
11
17
32
/;
```

```perl
my $input = get_challenge('2025/day/5/input');
```

```perl
sub process_day5 {
    my ($input) = @_;
    my ($fresh_inputs, $ingredient_inputs) = split /\n\n/, $input;
    my @fresh_rows = grep $_, map s/\s+//gr, split /\s+/, $fresh_inputs;
    my @ingredient_rows = grep $_, map s/\s+//gr, split /\s+/, $ingredient_inputs;

    @fresh_rows = map [ split '-' ], @fresh_rows;
    my @fresh_ingredients = grep { my $ing = $_; first { $_->[0] <= $ing && $ing <= $_->[1] } @fresh_rows } @ingredient_rows;
    
    return scalar @fresh_ingredients
}

timer sub {
    say process_day5($practice_input);
    confirm '3', process_day5($practice_input);
}
```

```perl
my $res1;
timer sub {
    $res1 = process_day5($input);
    say $res1;
    confirm '770', $res1;
}
```

```perl
say post_answer('2025/day/5/answer', 1, $res1); # 770
```

```perl
sub intersects {
    my ($a, $b) = @_;
    return !($a->[0] > $b->[1]
        || $a->[1] < $b->[0])
}
sub combine {
    my @inputs = @_;
    return [
        (min map { $_->[0] } @inputs),
        max map { $_->[1] } @inputs
    ]
}
sub process_day5_part2 {
    my ($input) = @_;
    my ($fresh_inputs, $ingredient_inputs) = split /\n\n/, $input;
    my @fresh_rows = grep $_, map s/\s+//gr, split /\s+/, $fresh_inputs;

    my @rows = map [ split '-' ], @fresh_rows;
    @rows = sort { $a->[0] <=> $b->[0] } @rows; 
    my @colors = 0 .. $#rows;
    # say 'row: ', Dumper \@rows;
    foreach my $i (0 .. $#rows) {
        timecheck;
        my @overlap_indexs = grep intersects($rows[$i], $rows[$_]), ($i+1) .. $#rows;
        # say '@overlap_indexs:', Dumper \@overlap_indexs;
        foreach (@overlap_indexs) {
            $colors[$_] = $colors[$i];
        }
    }
    say 'colors: ', join ',', @colors;

    my %row_groups =
        group_by sub { $_->{color} },
        map { { color => $colors[$_], row => $rows[$_] } }
        0 .. $#rows;
    my @combined_rows = map { combine(map $_->{row}, @$_) } values %row_groups;
    # say Dumper \@combined_rows;
    
    sum map { $_->[1] - $_->[0] + 1 } @combined_rows
}

timer sub {
    say process_day5_part2($practice_input);
    say process_day5_part2(q/
    1-10
    5-6
    7-8
    21-22
    10-20
    0-1
    90-90

    asdf
    /);
    confirm '14', process_day5_part2($practice_input);
}
```

```perl
my $res2;
timer sub {
    $res2 = process_day5_part2($input);
    say $res2;
    confirm '357674099117260', $res2;
}
```

```perl
say post_answer('2025/day/5/answer', 2, $res2); # 357674099117260
```
