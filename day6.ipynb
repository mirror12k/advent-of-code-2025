```perl
sub process_day6 {
    my ($input) = @_;
    my $rows = transpose_2d parse_2d_string_array $input;
    # say 'debug: ', Dumper $rows;
    
    my %sign_subs = (
        '+' => sub { $_[0] + $_[1] },
        '-' => sub { $_[0] - $_[1] },
        '*' => sub { $_[0] * $_[1] },
    );
    
    my %sign_start_values = (
        '+' => 0,
        '-' => 0,
        '*' => 1,
    );

    my @reduced_rows = @{ map_rows sub {
        my $sign = $_->[-1];
        reduce sub { $sign_subs{$sign}->(@_) }, $sign_start_values{$sign}, @$_[0 .. @$_ - 2]
    }, $rows };
    # say 'reduced_rows: ', Dumper \@reduced_rows;
    sum @reduced_rows
    # my @rows = map [ grep $_, split /\s+/ ], grep $_, split /\n/, $input;
    # my $signs = pop @rows;
    # @rows
}

timer sub {
    say Dumper process_day6($practice_input);
    confirm '4277556', process_day6($practice_input);
}
```

```perl
sub group_spaced {
    my ($spacer, @vals) = @_;
    my @groups = ([]);
    foreach (@vals) {
        if ($_ ne $spacer) { push @{$groups[-1]}, $_ } else { push @groups, [] }
    }
    return @groups;
}
sub assert { my ($a, $b) = @_; die "assertion failed: $a != $b" unless $a eq $b; return $a }
```

```perl
sub process_day6_part2 {
    my ($input) = @_;
    my $rows = parse_2d_map_array $input;
    my @signs = grep $_ ne ' ', @{ pop @$rows };
    my @column_numbers = @{ map_rows sub { (join '', @$_) =~ s/\s+//gr }, transpose_2d $rows };
    my @column_groups = group_spaced '', @column_numbers;
    assert scalar(@signs), scalar(@column_groups);
    # say Dumper \@column_groups;

    my %sign_subs = (
        '+' => sub { $_[0] + $_[1] },
        '-' => sub { $_[0] - $_[1] },
        '*' => sub { $_[0] * $_[1] },
    );
    
    my %sign_start_values = (
        '+' => 0,
        '-' => 0,
        '*' => 1,
    );

    my @reduced_cols = map {
        my ($sign, $nums) = ($signs[$_], $column_groups[$_]);
        reduce sub { $sign_subs{$sign}->(@_) }, $sign_start_values{$sign}, @$nums
    } 0 .. $#signs;
    sum @reduced_cols
    
}

timer sub {
    say Dumper process_day6_part2($practice_input);
    confirm '3263827', process_day6_part2($practice_input);
}
```
